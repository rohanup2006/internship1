```python
import pandas as pd
one = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 40],
    'id':[1,2,3,4]
})
two = pd.DataFrame({
    'name': ['David', 'Eve', 'Frank', 'Grace'],
    'marks': [80, 85, 90, 95],
    'id':[1,2,3,5]
})
left_join = pd.merge(one, two,on = 'id', how='left')
print("Left Join:")
print(left_join)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Department': ['HR', 'IT', 'Finance']
})
df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')
index_join = df1_index.join(df2_index, how='inner')
print("Index Join:")
print(index_join)
```

```python
import pandas as pd
one = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 40],
    'id':[1,2,3,4]
})
two = pd.DataFrame({
    'name': ['David', 'Eve', 'Frank', 'Grace'],
    'marks': [80, 85, 90, 95],
    'id':[1,2,3,5]
})
right_join = pd.merge(one, two, on='id', how='right')
print("Right Join:")
print(right_join)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance']
}, index=[2, 3, 4])
result = df1.join(df2, how='inner')

print(result)
```

```python
df3 = pd.DataFrame({
    'ID': [1, 1, 2, 2],
    'Year': [2023, 2024, 2023, 2024],
    'Sales': [100, 150, 200, 250]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 2],
    'Year': [2023, 2023, 2024],
    'Profit': [20, 40, 60]
})

multi_merge = pd.merge(df3, df4, on=['ID','Year'], how='inner')

print(multi_merge)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Charlie', 'David']
})

df3 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Score': [85, 90, 88, 92]
})
print(df1)
print(df2)
print(df3)
```

```python
concat_df = pd.concat([df1, df2], axis=0)

print(concat_df)
```

```python
final_df = pd.merge(concat_df, df3, on='ID', how='inner')

print(final_df)
```
