```python
num1 = int(input("enter first element:"))
num2 = int(input("enter second element:"))
add = num1 + num2
subtraction = num1 - num2
multiply = num1 * num2
if num2 != 0:
    divide = num1 / num2
else:
    divide = "not allowed"    

print("Addition:", add)
print("Subtraction:", subtraction)
print("Multiplication:", multiply)
print("Division:", divide)
```
