```python
import pandas as pd
data = {
    "Name": ["Rohan", "Rohit", "Aman"],
    "Marks": [85, 90, 88]
}
df = pd.DataFrame(data)
new_row = {"Name": "Sohan", "Marks": 92}
df = pd.concat([df.iloc[:1], pd.DataFrame([new_row]), df.iloc[1:]]).reset_index(drop=True)
print(df)
```
